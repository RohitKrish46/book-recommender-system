```python
import pandas as pd
import numpy as np
```

```python
books_df = pd.read_csv('../data/BX-Books.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
```

```python
books_df.info()
```

```python
books_df = books_df[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
books_df.head()
```

```python
books_df.rename(columns={"Book-Title": 'title',
                         "Book-Author": 'author',
                         "Year-Of-Publication": 'year',
                         "Publisher": 'publisher',
                         "Image-URL-L": 'image_url'}, inplace=True)
books_df.head()
```

```python
users_df = pd.read_csv('../data/BX-Users.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
```

```python
users_df.head()
```

```python
users_df.info()
```

```python
users_df.rename(columns={"User-ID": "user_id",
                         "Location": "location",
                         "Age": "age"}, inplace=True)
users_df.head()
```

```python
ratings_df = pd.read_csv('../data/BX-Book-Ratings.csv', sep=';', on_bad_lines='skip', encoding='latin-1')
```

```python
ratings_df.head()
```

```python
ratings_df.rename(columns={"User-ID":'user_id',
                      'Book-Rating':'rating'},inplace=True)
ratings_df.head()
```

```python
ratings_df['user_id'].value_counts()
```

```python
# storing users who has give above 200 ratings
x = ratings_df['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
ratings_df = ratings_df[ratings_df['user_id'].isin(x[x].index)]
ratings_df.head()
```

```python
# merge books and ratings
ratings_with_books = ratings_df.merge(books_df, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
num_of_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_of_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
num_of_rating.head()
```

```python
final_ratings = ratings_with_books.merge(num_of_rating, on='title')
final_ratings.head()
```

```python
# Lets take those books which got at least 50 rating of user

final_ratings = final_ratings[final_ratings['num_of_rating'] >= 50]
final_ratings.head()
```

```python
final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)
final_ratings.shape
```

```python
final_ratings.isnull().sum()
```

```python
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

book_sparse = csr_matrix(book_pivot)

model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_names = book_pivot.index
```

```python
np.where(book_pivot.index == '4 Blondes')[0][0]
ids = np.where(final_ratings['title'] == "4 Blondes")[0][0]
final_ratings.iloc[ids]['image_url']
```

```python
book_name = []
ids_index = []
for book_id in suggestion:
    book_name.append(book_pivot.index[book_id])
    
for name in book_name[0]: 
    ids = np.where(final_ratings['title'] == name)[0][0]
    ids_index.append(ids)

for idx in ids_index:
    url = final_ratings.iloc[idx]['image_url']
    print(url)
```

```python
import pickle
pickle.dump(model,open('../artifacts/model.pkl','wb'))
pickle.dump(book_names,open('../artifacts/book_names.pkl','wb'))
pickle.dump(final_ratings,open('../artifacts/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('../artifacts/book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggested books are: \n")
                else:
                    print(j)
```
